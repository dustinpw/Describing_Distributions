```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv('Data/data (6).csv')
```

```python
df.info()
```

```python
VitD_levels = df['VitD_levels']
Doc_visits = df['Doc_visits']
TotalCharge = df['TotalCharge']
```

```python
# Plotting VitD_levels
plt.figure(figsize=(8, 6))
sns.histplot(VitD_levels, kde=True)
plt.axvline(np.mean(VitD_levels), color='red', linestyle='--', label='Mean')
plt.axvline(np.median(VitD_levels), color='green', linestyle='--', label='Median')
plt.axvline(np.mean(VitD_levels) + np.std(VitD_levels), color='black', linestyle='--', label='+1 Std')
plt.axvline(np.mean(VitD_levels) - np.std(VitD_levels), color='black', linestyle='--', label='-1 Std')
plt.axvspan(np.mean(VitD_levels) - np.std(VitD_levels), np.mean(VitD_levels) + np.std(VitD_levels), color='yellow', alpha=0.3, label='1 Std Range')
plt.legend()
plt.xlabel('VitD_levels')
plt.ylabel('Frequency')
plt.title('Histogram with KDE - VitD_levels')
plt.show()
```

```python
from scipy import stats
kurt = stats.kurtosis(VitD_levels, fisher = False)
kurt
```

- <b> Is it Discrete or Continuous?
    - continuous
    
- <b> Does it have a skew? If so, which direction (+/-)
    - No skew
    
- <b> What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)
    - Mesokurtic
